```python
# import required libs
import os
import pickle
import random
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
# import dataset
DIRECTORY = r'C:\Users\EED\Desktop\cat-dog-cnn\catsdogs\dataset'
CATEGORIES = ['cat', 'dog']
```

```python
# labeling of the data i.e. cat or dog
# it will take some mintues because data is large
IMG_SIZE = 96

data = []

for category in CATEGORIES:
    folder = os.path.join(DIRECTORY, category)
    label = CATEGORIES.index(category)
    for img in os.listdir(folder):
        img_path = os.path.join(folder, img)
        img_arr = cv.imread(img_path)
        img_arr = cv.resize(img_arr, (IMG_SIZE, IMG_SIZE))
        data.append([img_arr, label])
```

```python
# check data lenght
len(data)
```

```python
data
```

```python
# now shuhffle the data for better results
random.shuffle(data)
```

```python
# creating a list X and y for features and labels

X = []
y = []

for features, label in data:
    X.append(features)
    y.append(label)
```

```python
# convert X and y into array
X = np.array(X)
y = np.array(y)
```
